```python
import numpy as np

def sigmoid(z):
    return 1 / (1 + np.exp(-z))

def compute_gradient_reg(X, y, w, lambda_):
    m, n = X.shape

    # Compute predictions
    z = np.dot(X, w)
    f = sigmoid(z)

    # Compute gradients
    error = f - y
    gradient = (1 / m) * np.dot(X.T, error)

    # Regularize gradient (exclude bias term)
    gradient[1:] += (lambda_ / m) * w[1:]

    return gradient
```

```python
def gradient_descent(X, y, w, alpha, lambda_, num_iters):
    m = X.shape[0]

    for _ in range(num_iters):
        # Compute gradient
        grad = compute_gradient_reg(X, y, w, lambda_)

        # Update weights
        w -= alpha * grad

    return w
```

```python
# Example data (including bias term)
X = np.array([[1, 0.5, 1.5], [1, 1, 1], [1, 1.5, 0.5]])
y = np.array([0, 0, 1])
w = np.zeros(X.shape[1])  # Initialize weights
alpha = 0.01  # Learning rate
lambda_ = 1.0  # Regularization parameter
num_iters = 1000  # Number of iterations

# Perform gradient descent
w_opt = gradient_descent(X, y, w, alpha, lambda_, num_iters)

print(f'Optimized weights: {w_opt}')
```
