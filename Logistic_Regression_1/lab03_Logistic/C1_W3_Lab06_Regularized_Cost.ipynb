```python
import numpy as np

def sigmoid(z):
    return 1 / (1 + np.exp(-z))

def compute_cost_reg(X, y, w, lambda_):
    m = X.shape[0]

    # Compute predictions
    z = np.dot(X, w)  # Ensure shapes match
    f = sigmoid(z)

    # Compute the cost with regularization
    cost = (-1 / m) * (np.dot(y, np.log(f)) + np.dot((1 - y), np.log(1 - f)))
    reg_cost = cost + (lambda_ / (2 * m)) * np.sum(np.square(w[1:]))

    return reg_cost

# Example data (assuming X has a column of ones)
X = np.array([[1, 1, 2], [1, 3, 4], [1, 5, 6]])  # Add bias term (column of ones)
y = np.array([0, 1, 1])
w = np.array([0.1, 0.2, 0.3])  # Including bias term
lambda_ = 1.0  # Regularization parameter

# Compute the cost
cost = compute_cost_reg(X, y, w, lambda_)
print(f'Regularized Cost: {cost:.4f}')
```
